```python
import pandas as pd
import numpy as np
```

```python
cf_df = pd.read_excel('/content/drive/MyDrive/CTCP Vận tải và xếp dỡ Hải An/Lưu chuyển tiền tệ/HAH_cf.xlsx')
cf_df
```

```python
#Đổi kiểu dữ liệu thành int
for cells in cf_df.iloc[:,2:]:
  cf_df[cells] = pd.to_numeric(cf_df[cells])
#fill NaN bằng 0
cf_df = cf_df.fillna(0)
cf_df
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame
cf_df_unpivoted = pd.melt(cf_df, id_vars=['Code', 'Chỉ tiêu'], var_name='Quý', value_name='Giá trị')

# Display the unpivoted table
cf_df_unpivoted
```

```python
import datetime
from datetime import datetime
def convert_quarter_to_date(quarter_str):
    # Mapping quarters to their end dates
    quarter_end_dates = {
        "Quý 1": "03-31",
        "Quý 2": "06-30",
        "Quý 3": "09-30",
        "Quý 4": "12-31"
    }

    # Split the input string to get the quarter and year
    quarter, year = quarter_str.split('- ')

    # Get the end date for the given quarter
    end_date = quarter_end_dates[quarter]

    # Combine year and end date to form the final date string
    final_date_str = f"{year}-{end_date}"

    # Convert the final date string to a datetime object
    final_date = datetime.strptime(final_date_str, "%Y-%m-%d")

    return final_date
cf_df_unpivoted['Date'] = cf_df_unpivoted['Quý'].apply(convert_quarter_to_date)
#Xóa cột Quý
cf_df_unpivoted = cf_df_unpivoted.drop(columns=['Quý'])
cf_df_unpivoted
```

```python
#Lưu vào file excel
cf_df_unpivoted.to_excel('/content/drive/MyDrive/CTCP Vận tải và xếp dỡ Hải An/Lưu chuyển tiền tệ/cf_df_unpivoted.xlsx', index=)
```
